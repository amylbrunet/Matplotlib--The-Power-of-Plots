## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "Data/Mouse_metadata.csv"
study_results_path = "Data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
merged_data_df = pd.merge(mouse_metadata, study_results, how='outer', on='Mouse ID')

# Display the data table for preview
merged_data_df
```

```python
# Checking the number of mice.
merged_data_df['Mouse ID'].value_counts()
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
# Optional: Get all the data for the duplicate mouse ID. 
duplicate_df = merged_data_df.set_index('Mouse ID')
duplicate_df.loc['g989']

```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_df = duplicate_df.drop(['g989'])
```

```python
# Checking the number of mice in the clean DataFrame.
new_clean_df = clean_df.reset_index()
new_clean_df['Mouse ID'].value_counts()
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
tumor_volume_stats = new_clean_df.groupby(['Drug Regimen'])

# mean, median, variance, standard deviation, and SEM of the tumor volume. 
tumor_mean = tumor_volume_stats['Tumor Volume (mm3)'].mean()
tumor_med = tumor_volume_stats['Tumor Volume (mm3)'].median()
tumor_var = tumor_volume_stats['Tumor Volume (mm3)'].var()
tumor_std = tumor_volume_stats['Tumor Volume (mm3)'].std()
tumor_sem = tumor_volume_stats['Tumor Volume (mm3)'].sem()

# Assemble the resulting series into a single summary dataframe.
tumor_volume_stats_df = pd.DataFrame({"Mean": tumor_mean,
                                     "Median" : tumor_med,
                                     "Variance" : tumor_var,
                                     "Standard Deviation" : tumor_std,
                                     "SEM" : tumor_sem})
tumor_volume_stats_df

```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Using the aggregation method, produce the same summary statistics in a single line
tumor_stats_2 = new_clean_df.groupby(['Drug Regimen'])['Tumor Volume (mm3)'].agg(['mean', 'median', 'var', 'std', 'sem'])
tumor_stats_2
```

## Bar and Pie Charts

```python
# Generate a bar plot showing the total number of timepoints for all mice tested for each drug regimen using Pandas.
timepoint_bar = new_clean_df['Drug Regimen'].value_counts().plot.bar()

#create labels for axes
timepoint_bar.set_xlabel('Drug Regimen')
timepoint_bar.set_ylabel('Number of Timepoints')
timepoint_bar.set_title('Number of Timepoints for Each Drug Regimen')

#show the bar plot
plt.show()
```

```python
# Generate a bar plot showing the total number of timepoints for all mice tested for each drug regimen using pyplot.

# calculate amount of timepoints for y axis
timepoint_count = new_clean_df['Drug Regimen'].value_counts()

# set x axis for the length of the time point count
x_axis = np.arange(len(timepoint_count))

# create bar plot with x axis, y axis, and width to match bar plot above
plt.bar(x_axis, timepoint_count, width = 0.5)

#label the x axis tick marks
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, timepoint_count.index.values)

# rotate x axis labels for clarity
plt.xticks(rotation=90)

# create labels for axes
plt.xlabel('Drug Regimen')
plt.ylabel('Number of Timepoints')
plt.title('Number of Timepoints for Each Drug Regimen')

# show bar plot
plt.show()
```
